```python
#AFFINE TRANSFORMATON
path="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\4.1.03.tiff"
#imgpath="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\4.1.04.tiff"
#img1=cv2.imread(imgpath,1)
img=cv2.imread(path,1)
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
rows,columns,channels=img1.shape
point1=np.float32([[100,100],[300,100],[100,200]])
point2=np.float32([[200,150],[400,150],[200,400]])
R= cv2.getAffineTransform(point1,point2)
#trans=np.float32([[1,0,5],[0,1,-50]])
op=cv2.warpAffine(img1,R,(columns,rows))
cv2.imshow("OUTPUT",op)
cv2.waitKey(0)
cv2.destroyAllWindows()
#cv2.imshow("Monkey",img)
#cv2.imwrite(outpath,img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#PERSPECTIVE TRANSFORMATON
path="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\4.1.03.tiff"
#imgpath="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\4.1.04.tiff"
#img1=cv2.imread(imgpath,1)
img=cv2.imread(path,1)
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
rows,columns,channels=img1.shape
point1=np.float32([[0,0],[400,0],[0,400],[400,400]])
point2=np.float32([[0,0],[300,0],[0,300],[300,300]])
R= cv2.getPerspectiveTransform(point1,point2)
#trans=np.float32([[1,0,5],[0,1,-50]])
op=cv2.warpPerspective(img1,R,(columns,rows))
cv2.imshow("OUTPUT",op)
cv2.waitKey(0)
cv2.destroyAllWindows()
#cv2.imshow("Monkey",img)
#cv2.imwrite(outpath,img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#Segmentation
path="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\SONU.jpg"
#imgpath="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\america.jpeg"
#img1=cv2.imread(imgpath,1)
img=cv2.imread(path,1)
th=127
max_val=255
#img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,o1= cv2.threshold(img,th,max_val,cv2.THRESH_BINARY)
ret,o2= cv2.threshold(img,th,max_val,cv2.THRESH_BINARY_INV)
ret,o3= cv2.threshold(img,th,max_val,cv2.THRESH_TOZERO)
ret,o4= cv2.threshold(img,th,max_val,cv2.THRESH_TOZERO_INV)
ret,o5= cv2.threshold(img,th,max_val,cv2.THRESH_TRUNC)
op=[img,o1,o2,o3,o4,o5]
titles=["Original","Binary","Biary Inverted","Tozero","Tozer_inverted","Truncation"]
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(op[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
#cv2.imshow("OUTPUT",op)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
#cv2.imshow("Monkey",img)
#cv2.imwrite(outpath,img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#Otsu's Binarization Thresholding
path="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\SONU.jpg"
#imgpath="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\america.jpeg"
#img1=cv2.imread(imgpath,1)
img=cv2.imread(path,0)
th=127
max_val=255
#img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,o1= cv2.threshold(img,th,max_val,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret,o2= cv2.threshold(img,th,max_val,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
ret,o3= cv2.threshold(img,th,max_val,cv2.THRESH_TOZERO+cv2.THRESH_OTSU)
ret,o4= cv2.threshold(img,th,max_val,cv2.THRESH_TOZERO_INV+cv2.THRESH_OTSU)
ret,o5= cv2.threshold(img,th,max_val,cv2.THRESH_TRUNC+cv2.THRESH_OTSU)
op=[img,o1,o2,o3,o4,o5]
titles=["Original","Binary","Biary Inverted","Tozero","Tozer_inverted","Truncation"]
for i in range(6):
    plt.imshow(op[i],cmap="gray")
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
#cv2.imshow("OUTPUT",op)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
#cv2.imshow("Monkey",img)
#cv2.imwrite(outpath,img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#Adaptive Thresholding
path="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\SONU.jpg"
#imgpath="C:\\Users\\ARIJIT ROYCHAUDHURY\\Downloads\\misc\\misc\\america.jpeg"
#img1=cv2.imread(imgpath,1)
img=cv2.imread(path,0)
block_size=513
c=12
th1=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,block_size,c)
th2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,block_size,c)
op=[img,th1,th2]
titles=["Original","Mean Adaptive","Gaussian Adaptive"]
for i in range(3):
    plt.imshow(op[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
    plt.show()
#cv2.imshow("OUTPUT",op)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
#cv2.imshow("Monkey",img)
#cv2.imwrite(outpath,img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
